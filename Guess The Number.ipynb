```python
# importing required modules
import random
import tkinter as tk
from tkinter import messagebox
from PIL import Image , ImageTk
import math

def check_min_or_max(num,guess_num):
    if num<guess_num:
        return "You Guessed Too Large"
    else:
        return "You Gussed Too Small"
    
def multiple(num,guess_num):
    return str(num*random.choice(list(range(2,10))))+" is multiple of \n Original Number"
    
def divisible(num,guess_num):
    li = []
    for i in range(2,int(math.sqrt(num)+1)):
        if num%i==0:
            li.append(i)
    
    if len(li)==0:
        return "Number is divisible by 1"
    
    return "Number is divisible by "+str(random.choice(li))

def is_prime(num,guess_num):
    flag = False
    for i in range(2,int(math.sqrt(num)+1)):
        if num%i==0:
            flag = True
            break
    if flag:
        return "Number is Prime"
    return "Number is not Prime"
        

l = [check_min_or_max , multiple , divisible , is_prime]
life = 5
i = -1

def checking(screen , num , guess_num , tries , hint , submit):
    global l
    global life
    life = life - 1
    global i
    i+=1
    if num==guess_num:
        hint.configure(text="Congratulations ! \n You Guessed Right Number")
        submit.configure(text="Play Again" , command=lambda : start(screen))
    elif life==0:
        tries.configure(text="Tries Left : "+str(life))
        hint.configure(text="You Gussed Wrong Number")
        submit.configure(text="Play Again" , command=lambda : start(screen))
    else:
        hint.configure(text=l[i](num,guess_num))
        tries.configure(text="Tries Left : "+str(life))

        hint.configure(text=l[i](num,guess_num))
        tries.configure(text="Tries Left : "+str(life))
    

def guessing(screen , s , e , life):
    
    num = random.randint(s,e)
    
    label = tk.Label(screen , bg = "indigo",width=600,height=500)
    label.place(x=0,y=0)
    
    tries = tk.Label(screen , bg="indigo",text="Tries Left : "+str(life),width=25,font=("Ariel",30,'bold'))
    tries.place(x=0,y=0)
    
    guess = tk.Label(screen , bg="indigo",text="Guess The Number",width=25,font=("Ariel",30,'bold'))
    guess.place(x=0,y=150)
    
        
    guess_num = tk.StringVar()

    enter = tk.Entry(screen , bg="indigo" , textvariable=guess_num , width = 26 , font=('Ariel',30,'bold'))
    enter.place(x=10,y=200)
    
    submit = tk.Button(screen , text="Submit" ,bg="blue" , width=19 , font=("Ariel",40,'bold') , command= lambda : checking(screen , num , int(guess_num.get()) , tries , hint , submit))
    submit.place(x=0,y=400)
    
    hint = tk.Label(screen , bg='indigo' , text="First Time ! No Hint" , width=25,font=('Ariel',30,"bold"))
    hint.place(x=0,y=300)
    
    

def playing(screen,s,e , life):
        
    try:
        s = int(s)
        e = int(e)
        if s>e:
            raise Exception()
        else:
            messagebox.showinfo("Information","You have 5 tries in this game. Guess the correct number within the given tries")
            guessing(screen , s , e , life)
    except:
        messagebox.showwarning("Warning","Range not valid")


def start(screen):
    global life , i
    life = 5
    i = -1
    label = tk.Label(screen , bg = "blue",width=600,height=500)
    label.place(x=0,y=0)
    
    select_range = tk.Label(screen , bg="blue",text="Select Range",width=25,font=("Ariel",30,'bold'))
    select_range.place(x=0,y=0)
    
    fr = tk.Label(screen , bg='blue',text="FROM",width=25,font=("Ariel",30,'bold'))
    fr.place(x=0,y=100)
    s = tk.StringVar()
    fre = tk.Entry(screen,bg='blue',textvariable=s,width=26,font=('Ariel',30,'bold'))
    fre.place(x=10,y=150)
    
    to = tk.Label(screen , bg='blue',text="TO",width=25,font=("Ariel",30,'bold'))
    to.place(x=0,y=200)
    e = tk.StringVar()
    toe = tk.Entry(screen,bg='blue',textvariable=e,width=26,font=('Ariel',30,'bold'))
    toe.place(x=10,y=250)
    
    
    play = tk.Button(screen , text="PLAY" , font=("Ariel",40,'bold'),fg="white",bg="indigo",width=19,command=lambda : playing(screen , s.get() , e.get() , 5))
    play.place(x=0,y=400)

# main function
def main():
    
    screen = tk.Tk()
    screen.title("Guess The Number")
    screen.geometry(("600x500"))
    screen.minsize(600 , 500)
    screen.maxsize(600 , 500)
    img = Image.open("guess.png")
    r_img = img.resize((600,400))
    r_img = ImageTk.PhotoImage(r_img)
    label = tk.Label(screen,image=r_img)
    label.place(x=0,y=0)
    
    play = tk.Button(screen , text="START" , font=("Ariel",40,'bold'),fg="white",bg="green",width=19,command=lambda : start(screen))
    play.place(x=0,y=400)
    
    
    screen.mainloop()
    
    

# calling main function
if __name__=="__main__":
    main()
```
